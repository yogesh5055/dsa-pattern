11.You are given an integer N and the first 6 integers of a sequence.
Generate a sequence of length N using the following rules:

The first 6 elements of the sequence are given as input.

For every element from index 6 onwards, each element is the sum of the previous two elements.

Print the generated sequence up to N elements.

ðŸ“¥ Sample Input
10
1 2 3 4 5 6

ðŸ“¤ Sample Output
1 2 3 4 5 6 11 17 28 45

```python
n = int(input())
arr = list(map(int, input().split()))
if n <= 6:
    print(*arr[:n])
else:
    for i in range(6, n):
        arr.append(arr[i-1] + arr[i-2])

    print(*arr)
```

12.You are given an integer N and an array A of size N.
The array is considered circular.

For each element in the array, count how many of its immediate neighbors are greater than it.

Each element has two neighbors:

Left neighbor

Right neighbor

Since the array is circular:

The left neighbor of the first element is the last element

The right neighbor of the last element is the first element

Print the count for each element in the array.

ðŸ“¥ Sample Input
5
1 3 2 5 4

ðŸ“¤ Sample Output
2 0 2 0 1

```python
n = int(input())
arr = list(map(int, input().split()))

res = []

for i in range(n):
    c = 0
    if arr[(i - 1) % n] > arr[i]:
        c += 1
    if arr[(i + 1) % n] > arr[i]:
        c += 1
    res.append(c)

print(*res)

```

13.You are given a string S.
Your task is to:

Count the frequency of each character in the string.

Among the characters with odd frequencies, find the maximum odd frequency.

Among the characters with even frequencies, find the minimum even frequency.

Print the absolute difference between:

maximum odd frequency

minimum even frequency

ðŸ“¥ Sample Input
aabbbbccddd

ðŸ“¤ Sample Output
2

```python
n = input().strip()
f = {}

# frequency count
for ch in n:
    f[ch] = f.get(ch, 0) + 1

max_odd = float('-inf')
min_even = float('inf')

for v in f.values():
    if v % 2 != 0:
        max_odd = max(max_odd, v)
    else:
        min_even = min(min_even, v)

print(abs(max_odd - min_even))
```

14.You are given an integer N and a list L of N integers.

Your task is to find the smallest positive integer in the list that occurs exactly once.

If no such positive integer exists, print -1.

ðŸ“¥ Sample Input
6
3 4 2 3 5 2

ðŸ“¤ Sample Output
4

```python
n = int(input())
arr = list(map(int, input().split()))

freq = {}

# count frequency
for x in arr:
    freq[x] = freq.get(x, 0) + 1

ans = float('inf')

# find smallest positive number with frequency 1
for k, v in freq.items():
    if k > 0 and v == 1:
        ans = min(ans, k)

print(ans if ans != float('inf') else -1)
```

15.You are given an integer N and a string S of length N.

Your task is to:

Count the frequency of each character in the string.

Then count the frequency of those frequencies (i.e., how many characters have the same frequency).

Find the frequency value that occurs maximum number of times.

If there is a tie (same maximum count), return the smallest frequency value.

Return that frequency.

ðŸ“¥ Sample Input
20
ACABDDABDCDACFAEGFDA

ðŸ“¤ Sample Output
1

```python
s = input().strip()

# step 1: character frequency
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

# step 2: frequency of frequency
freq_of_freq = {}
for v in freq.values():
    freq_of_freq[v] = freq_of_freq.get(v, 0) + 1

# step 3: find required frequency
max_count = -1
result = float('inf')

for frequency, count in freq_of_freq.items():
    if count > max_count or (count == max_count and frequency < result):
        max_count = count
        result = frequency

print(result)
```

16.You are given:

An integer N â†’ number of people to be transported

An integer P â†’ fuel units required per bus

An integer Q â†’ fuel units required per shuttle

Transportation details:

One bus can carry 80 people

One shuttle can carry 8 people

Fuel price is â‚¹75 per unit

Your task is to determine the minimum fuel cost required to transport all N people, by choosing an optimal combination of buses and shuttles.

You may use any number of buses and shuttles, but everyone must be transported.

ðŸ“¥ Sample Input
95
60 10


(Here: N = 95, P = 60, Q = 10)

ðŸ“¤ Sample Output
4500

```python
# n = int(input())
# p = int(input())
# q = int(input())

def minFuelCost(N, P, Q):
    busCap = 80
    shuttleCap = 8
    fuelPrice = 75

    minCost = float('inf')

    # try all possible number of buses
    maxBuses = (N + busCap - 1) // busCap

    for buses in range(maxBuses + 1):
        remaining = N - buses * busCap
        if remaining < 0:
            remaining = 0

        shuttles = (remaining + shuttleCap - 1) // shuttleCap
        cost = (buses * P + shuttles * Q) * fuelPrice

        minCost = min(minCost, cost)

    return minCost


# example usage
print(minFuelCost(95, 60, 10))

```

17.You are given:

An integer N â†’ number of buildings

An array H of size N, where H[i] represents the height of the i-th building

An integer D â†’ a fixed distance

Your task is to find the sum of heights of all buildings that satisfy the following condition:

For a building at index i:

If there exists a building at index i âˆ’ D, then H[i] must be greater than H[i âˆ’ D]

If there exists a building at index i + D, then H[i] must be greater than H[i + D]

If one of these indices is out of bounds, that side is considered valid automatically.

Add the heights of all such buildings and print the final sum.

ðŸ“¥ Sample Input
6
1 3 2 1 5 4
2

ðŸ“¤ Sample Output
8

```python
def sumTallBuildings(N, H, D):
    total = 0

    for i in range(N):
        left_ok = (i - D < 0) or (H[i] > H[i - D])
        right_ok = (i + D >= N) or (H[i] > H[i + D])

        if left_ok and right_ok:
            total += H[i]

    return total


# example usage
H = [1, 3, 2, 1, 5, 4]
D = 2
print(sumTallBuildings(len(H), H, D))
```

18.You are given a string S consisting of letters and digits.

Your task is to count how many digit-only substrings exist in the string such that:

The substring contains only digits

The digit substring is immediately preceded by a letter

The digit substring is immediately followed by a letter

Only such digit substrings should be counted.

ðŸ“¥ Sample Input
a123d

ðŸ“¤ Sample Output
1

```python
def countDigitSubstrings(S):
    count = 0
    i = 0
    n = len(S)

    while i < n:
        if S[i].isdigit():
            start = i
            while i < n and S[i].isdigit():
                i += 1
            end = i - 1

            if (start > 0 and end < n - 1 and
                S[start - 1].isalpha() and
                S[end + 1].isalpha()):
                count += 1
        else:
            i += 1

    return count


# example usage
S = "a123d"
print(countDigitSubstrings(S))
```

19.You are given a string S.

Your task is to replace every group of consecutive identical characters of length â‰¥ 2 with a single # symbol.

Rules:

If a character appears only once consecutively, keep it as it is.

If two or more identical characters appear consecutively, replace that entire group with one #.

If multiple such groups occur back-to-back, do not add multiple #, just keep one.

Return the final modified string.

ðŸ“¥ Sample Input
aabbbccdeeea

ðŸ“¤ Sample Output
#d#a

```python
def replaceConsecutive(S):
    res = []
    n = len(S)
    i = 0

    while i < n:
        j = i
        while j < n and S[j] == S[i]:
            j += 1

        # length of the consecutive group
        if j - i >= 2:
            if not res or res[-1] != '#':
                res.append('#')
        else:
            res.append(S[i])

        i = j

    return ''.join(res)


# example usage
S = "aabbbccdeeea"
print(replaceConsecutive(S))
```

20.You are given a string text consisting of lowercase English letters (a to z).

Each character corresponds to a position based on its alphabetical index:

a â†’ 0, b â†’ 1, c â†’ 2, ..., z â†’ 25


Your task is to calculate the total typing distance, defined as:

For every pair of adjacent characters in the string,
compute the absolute difference between their alphabetical indices.

Sum up all these differences.

Return the final total distance.

ðŸ“¥ Sample Input
cba

ðŸ“¤ Sample Output
2

```python
def totalTypingDistance(text):
    distance = 0
    for i in range(1, len(text)):
        prev_index = ord(text[i - 1]) - ord('a')
        curr_index = ord(text[i]) - ord('a')
        distance += abs(curr_index - prev_index)
    return distance
text = "cba"
print(totalTypingDistance(text))
```
