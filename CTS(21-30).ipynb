21.You are given a string S consisting only of the characters 'X' and 'Y'.

A string is called balanced if:

Characters at even indices and odd indices alternate between X and Y

That means the string should be in either of the following forms:

Starts with X â†’ X Y X Y X Y ...

Starts with Y â†’ Y X Y X Y X ...

In one conversion, you can change any character (X to Y or Y to X).

ðŸ‘‰ Task:
Find the minimum number of conversions required to make the given string balanced.

ðŸ“¥ Sample Input
XXYXYY

ðŸ“¤ Sample Output
2

```python
def min_conversions_to_balanced(s):
    a = 0   # changes if pattern starts with X
    b = 0   # changes if pattern starts with Y

    for i in range(len(s)):

        # pattern 1: X Y X Y ...
        if i % 2 == 0 and s[i] != 'X':
            a += 1
        if i % 2 == 1 and s[i] != 'Y':
            a += 1

        # pattern 2: Y X Y X ...
        if i % 2 == 0 and s[i] != 'Y':
            b += 1
        if i % 2 == 1 and s[i] != 'X':
            b += 1

    return min(a, b)


s = "XXYXYY"
print(min_conversions_to_balanced(s))
```

22.You are given a string S representing product stock details.

Each product is separated by a semicolon (;)
Each product has the format:

ProductName:Price:Quantity


ðŸ‘‰ Task:
Calculate and return the total stock value, where:

Stock Value = Price Ã— Quantity


Add the stock value of all products and return the total.

ðŸ“¥ Sample Input
Apple:250:10;Banana:120:15;Orange:300:5

ðŸ“¤ Sample Output
5800

```python
def total_stock_value(s):
    products = s.split(";")
    total = 0

    for product in products:
        parts = product.split(":")
        if len(parts) >= 3:
            price = int(parts[1].strip())
            quantity = int(parts[2].strip())
            total += price * quantity

    return total


# Sample Input
s = "Apple:250:10;Banana:120:15;Orange:300:5"

# Output
print(total_stock_value(s))
```

24.You are given:

An integer n (number of registrations)

An array of strings registrations, each representing a date in dd-MM-yyyy format

A start date and an end date (also in dd-MM-yyyy format)

ðŸ‘‰ Task:
Count how many registration dates fall within the given date range (inclusive).

A date is valid if:

startDate â‰¤ registrationDate â‰¤ endDate

ðŸ“¥ Sample Input
n = 6
registrations = [
  "01-01-2023", "10-01-2023", "05-02-2023",
  "25-12-2022", "03-01-2023", "01-03-2023"
]
start = "01-01-2023"
end = "31-01-2023"

ðŸ“¤ Sample Output
3

```python
from datetime import datetime

def count_valid_registrations(n, registrations, start, end):
    fmt = "%d-%m-%Y"
    start_date = datetime.strptime(start, fmt)
    end_date = datetime.strptime(end, fmt)
    count = 0
    for reg in registrations:
        reg_date = datetime.strptime(reg, fmt)
        if start_date <= reg_date <= end_date:
            count += 1

    return count


# Sample Input
n = 6
registrations = [
    "01-01-2023", "10-01-2023", "05-02-2023",
    "25-12-2022", "03-01-2023", "01-03-2023"
]
start = "01-01-2023"
end = "31-01-2023"

# Output
print(count_valid_registrations(n, registrations, start, end))
```

25.You are given:

An integer n

An integer array a of size n

For each element in the array, count how many previous elements are smaller than it and get skipped until a larger or equal element blocks the process.

ðŸ‘‰ Task:
Calculate and return the total skip count for the entire array.

(The logic follows a stack-based comparison where smaller previous elements are removed when a larger element appears.)

ðŸ“¥ Sample Input
n = 6
a = [3, 1, 4, 2, 5, 1]

ðŸ“¤ Sample Output
4


```python

def calculate_total_skip_count(n, a):
    stack = []
    total_skip = 0

    for i in range(n):
        current_skip = 0

        while stack and a[i] > stack[-1]:
            stack.pop()
            current_skip += 1

        stack.append(a[i])
        total_skip += current_skip

    return total_skip
n = 6
a = [3, 1, 4, 2, 5, 1]

# Output
print(calculate_total_skip_count(n, a))
```

26.You are given:

An integer N (number of digits to consider)

An integer array A containing digits (0â€“9)

A string S consisting of uppercase English characters

For each character in S:

Find its ASCII value

Convert the ASCII value into digits

If any digit of the ASCII value is present in array A, count that character

ðŸ‘‰ Task:
Return the total count of such characters in the string S.

ðŸ“¥ Sample Input
N = 3
A = [1, 6, 4, 3, 6, 5]
S = "ABCDEF"

ðŸ“¤ Sample Output
6

```python
def count_chars(N, A, S):
    allowed = set(A)
    count = 0

    for ch in S:
        ascii_val = ord(ch)
        for digit in str(ascii_val):
            if int(digit) in allowed:
                count += 1
                break   

    return count



A = [1, 6, 4, 3, 6, 5]
S = "ABCDEF"


print(count_chars(3, A, S))
```

27.You are given a lowercase string s.

For each character in the string:

Find its alphabet position
(a = 1, b = 2, ..., z = 26)

If the position is a perfect square
(1, 4, 9, 16, 25)

Let k = âˆšposition

Check the adjacent characters (left or right):

If any adjacent characterâ€™s position equals k, keep the character unchanged

Otherwise, replace the character with the character whose position is k

If the position is not a perfect square, keep the character unchanged

ðŸ‘‰ Task:
Return the modified string.

ðŸ“¥ Sample Input
abcd

ðŸ“¤ Sample Output
abbd

```python
import math

def modify_string(s):
    result = []

    for i in range(len(s)):
        p = ord(s[i]) - ord('a') + 1

        # check if p is a perfect square
        if p in {1, 4, 9, 16, 25}:
            k = int(math.sqrt(p))

            left = i > 0 and (ord(s[i - 1]) - ord('a') + 1) == k
            right = i < len(s) - 1 and (ord(s[i + 1]) - ord('a') + 1) == k

            if left or right:
                result.append(s[i])
            else:
                result.append(chr(ord('a') + k - 1))
        else:
            result.append(s[i])

    return "".join(result)


# Sample Input
print(modify_string("abcd"))
```

28.You are given a string s containing digits and lowercase letters.

A character is called a wrapped item if:

The previous character is a digit

The current character is a lowercase letter

The next character is a digit

ðŸ‘‰ Task:
Count how many such wrapped items exist in the string.

ðŸ“¥ Sample Input
5aart6i7io8o5o56

ðŸ“¤ Sample Output
4

```python
def count_wrapped_items(s):
    count = 0

    for i in range(1, len(s) - 1):
        if s[i - 1].isdigit() and s[i].islower() and s[i + 1].isdigit():
            count += 1

    return count


# Sample Input
s = "5aart6i7io8o5o56"

# Output
print(count_wrapped_items(s))
```

29.You are given a string s.

The string is divided into layers as follows:

Layer 1 has 1 character

Layer 2 has 2 characters

Layer 3 has 3 characters

and so onâ€¦

For each layer:

If all characters in that layer are the same, it is called a homogeneous layer

ðŸ‘‰ Task:
Count how many homogeneous layers exist in the string.

Stop processing when there are not enough characters to form the next layer.

ðŸ“¥ Sample Input
aaabbbcccdddddd

ðŸ“¤ Sample Output
3

```python
def count_homogeneous_layers(s):
    count = 0
    pos = 0
    layer = 1
    n = len(s)

    while pos + layer <= n:
        first = s[pos]
        homogeneous = True

        for i in range(1, layer):
            if s[pos + i] != first:
                homogeneous = False
                break

        if homogeneous:
            count += 1

        pos += layer
        layer += 1

    return count


# Sample Input
s = "aaabbbcccdddddd"

# Output
print(count_homogeneous_layers(s))
```

30.You are given:

An integer N â†’ number of nails

An integer X â†’ speed factor

An integer array S of size N, where each element represents the size of a nail

There are two working modes:

ðŸ”¹ Mode 1: Single Mode

Each nail is processed independently

Time taken for one nail =

ceil(S[i] / X)


Total time = sum of time for all nails

ðŸ”¹ Mode 2: Dynamic Mode

All nails are processed together

Time taken = maximum nail size

ðŸ‘‰ Task:
Return the minimum time required by choosing the better of the two modes.

ðŸ“¥ Sample Input
N = 5
X = 4
S = [2, 2, 4, 1, 1]

ðŸ“¤ Sample Output
3

```python
def optimum_time(N, X, S):
    single_mode_time = 0
    max_nail_size = 0

    for size in S:
        # ceil(size / X)
        single_mode_time += (size + X - 1) // X

        if size > max_nail_size:
            max_nail_size = size

    dynamic_mode_time = max_nail_size

    return min(single_mode_time, dynamic_mode_time)


# Sample Input
N = 5
X = 4
S = [2, 2, 4, 1, 1]

# Output
print(optimum_time(N, X, S))
```
