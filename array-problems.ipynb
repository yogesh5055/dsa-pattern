```python
#1. Rotate the array N times 

def rotate_array(arr, n):
    if not arr:
        return arr
    n = n % len(arr)  # Handle cases where n is greater than array length
    return arr[-n:] + arr[:-n]

n = list(map(int, input("Enter array elements separated by space: ").split()))
k = int(input("Enter number of rotations: "))
rotated_array = rotate_array(n, k)
print("Rotated Array:", rotated_array)


```

```python
# Alternative method for rotation 

def rotate_array_alternative(arr, d):
     n = len(arr)
  
    # Repeat the rotation d times
     for i in range(d):
      
        # Left rotate the array by one position
        first = arr[0]
        for j in range(n - 1):
            arr[j] = arr[j + 1]
        arr[n - 1] = first
     return arr
n = list(map(int, input("Enter array elements separated by space: ").split()))
k = int(input("Enter number of rotations: "))
rotated_array = rotate_array(n, k)
print("Rotated Array:", rotated_array)
```
