2.Given an array A of size N, count the number of indices where the array can be split into two non-empty parts such that the absolute difference between the sum of the left part and the sum of the right part is even.

Sample Input:

N = 5  
A = [10, 10, 3, 7, 6]

Sample Output:

4

```python
n = int(input())
arr = list(map(int,input().split()))
total_sum = sum(arr)
prefix_sum = 0
count = 0
for i in range(len(arr) - 1):  
        prefix_sum += arr[i]
        suffix_sum = total_sum - prefix_sum

        if abs(prefix_sum - suffix_sum) % 2 == 0:
            count += 1
print(count)
```

3.Given an array of integers of size N, count the number of contiguous subarrays of length 3 such that the sum of the first and third elements is equal to the middle element.

Sample Input:

N = 8  
arr = [1, 2, 1, 3, 5, 2, 4, 2]


Sample Output:

3

```python
arr = list(map(int,input().split()))

result = []
for i in range(len(arr) - 2):
    sub = arr[i : i+3]
    result.append(sub)
count = 0

for i in range(len(result)):
    if result[i][0] + result[i][-1] == result[i][3//2]:
        count+=1
print(count)
```

4.You are given two uppercase strings S1 and S2. You can add characters to S1. Each added character has a cost of 1. Determine the minimum cost required to transform S1 into S2 such that for every character, the frequency in S1 becomes at least the frequency in S2.

Sample Input:

S1 = "ABD"  
S2 = "AABCCAD"


Sample Output:

4

```python
s1 = input().strip()
s2 = input().strip()

f = {}
for i in s2:
    f[i] = f.get(i,0)+1

for i in s1:
    f[i] -= 1
s = 0
for i in f.values():
    s+=i
print(s)
```

```python
# Alternative 
s1 = input().strip()
s2 = input().strip()

f = {}

# Count frequency of characters in s2
for ch in s2:
    f[ch] = f.get(ch, 0) + 1

# Reduce frequency using characters from s1
for ch in s1:
    if ch in f:
        f[ch] -= 1

# Sum only the positive values (needed characters)
cost = 0
for v in f.values():
    if v > 0:
        cost += v

print(cost)
```

5.Given an array of strings of size N, remove all vowels from each string. For each resulting string, find its length. Determine the maximum length obtained after vowel removal and return the factorial of that maximum length. If all strings become empty after removing vowels, return 0.

Sample Input:

N = 1  
arr = ["eio"]


Sample Output:

0

```python
arr=list(map(str,input().strip().split()))
print(arr)
v = ['a','e','i','o','u']
sol = []
for i in arr:
    ans = ''
    for j in i:
        if j.lower() not in v:
             ans+=j.lower()
    sol.append(ans)
ma_le = 0 
for i in sol:
    print(len(i))
    if len(i) > ma_le:
        ma_le = len(i)
print(ma_le)
fin = 1
for i in range(1,ma_le+1):
    fin*=i
print(fin)
```

6.Given a string S, count the number of consonants present in it. Return the factorial of the number of consonants. If the string contains no consonants, return 0.

Sample Input:

S = "ABC"


Sample Output:

2

```python
arr = list(input().strip())
print(arr)
v = ['a','e','i','o','u']
count = 0
for i in arr:
    if i.lower() not in v:
        count+=1
ans = 1
for i in range(1,count+1):
    ans*=i
print(ans)
```

8.Given a sentence S consisting of words separated by spaces, form a pair for each word using its first and last character. Count the frequency of each such pair and return the most frequent pair. If there is a tie, return the pair that appears first in the sentence.

Sample Input:

hello hi how are you


Sample Output:

ho

```python
words = list(map(str,input().strip().split()))
freq = {}  
for word in words:
        if len(word) > 0:
            pair = word[0] + word[-1]
            if pair in freq:
                freq[pair] += 1
            else:
                freq[pair] = 1
print(freq)
max_freq = 0
for v in freq.values():
        if v > max_freq:
            max_freq = v
for k in freq:
    if freq[k] == max_freq:
         print(k)
```

9.Given a numeric string S, where each number from 1 to 26 can be mapped to letters A to Z, count the total number of valid single-digit and two-digit mappings present in the string. A mapping is valid if the number formed lies between 1 and 26 (inclusive).

Sample Input:

S = "123"


Sample Output:

5

```python
s = input()
n = len(s)
count = 0
for i in range(n):
        # check single digit
        num1 = int(s[i])
        if 1 <= num1 <= 26:
            count += 1

        # check two digits
        if i + 1 < n:
            num2 = int(s[i:i+2])
            if 1 <= num2 <= 26:
                count += 1
print(count)
```

10.You are given an integer array A of size N and two integers L and R. Starting with a dopamine level of 0, traverse the array from left to right.

If the current element lies in the range [L, R] (inclusive), increase the dopamine level by 1.

Otherwise, decrease the dopamine level by 1.

After processing all elements, determine the maximum and minimum dopamine levels reached at any point during the traversal. Return them as two space-separated values.

Sample Input:

N = 5
L = 3
R = 7
A = [2, 4, 6, 1, 8]


Sample Output:

2 -2

```python
n = int(input())
L = int(input())
R = int(input())
A = list(map(int,input().split()))
dopamine = 0
max_dopamine = float('-inf')
min_dopamine = float('inf')

for num in A:
    if L <= num <= R:
            dopamine += 1
    else:
            dopamine -= 1

    max_dopamine = max(max_dopamine, dopamine)
    min_dopamine = min(min_dopamine, dopamine)
print(max_dopamine, min_dopamine)
```
