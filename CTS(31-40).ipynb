31.You are given a string s consisting of lowercase English letters.

A harmonic pattern is defined as a substring that:

Starts with the characters "ab", and

Continues following a specific valid harmonic structure (as determined by the helper method getHarmonicLength).

Your task is to find the length of the largest harmonic pattern present in the given string.

If no harmonic pattern exists, return 0.

ðŸ“¥ Input Format

A single string s

ðŸ“¤ Output Format

An integer representing the maximum length of any harmonic pattern found in the string

ðŸ§ª Sample Input 1
ababa

âœ… Sample Output 1
5

```python
def find_largest_harmonic_pattern(s):
    max_len = 0
    n = len(s)

    for i in range(n - 1):
        if s[i] == 'a' and s[i + 1] == 'b':
            length = 2
            j = i + 2
            expected = 'a'

            while j < n and s[j] == expected:
                length += 1
                expected = 'b' if expected == 'a' else 'a'
                j += 1

            max_len = max(max_len, length)

    return max_len
```

32.You are given a positive integer N.

Your task is to repeatedly add the digits of the number until the result becomes a single-digit number.

Finally, print that single-digit result.

ðŸ” What does â€œrepeatedly add digitsâ€ mean?

Take the number

Split it into its digits

Add all the digits

If the result has more than one digit, repeat the process

Stop only when a single digit remains

ðŸŽ¯ Goal

ðŸ‘‰ Reduce the number to a single digit by repeatedly summing its digits.

ðŸ“¥ Input Format

A single integer N

ðŸ“¤ Output Format

A single-digit integer (final reduced value)

ðŸ§ª Example 1
Input
9875

Step-by-step:

9 + 8 + 7 + 5 = 29

2 + 9 = 11

1 + 1 = 2 âœ… (single digit)

Output
2

```python
n = int(input())

while n >= 10:
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    n = s

print(n)
```

Define a function F(N) such that:

F(0) = 1

F(1) = 1

For N â‰¥ 2, the value of F(N) is calculated using the formula:

F(N) = F(Nâˆ’1) + 7 Ã— F(Nâˆ’2) + âŒŠN / 4âŒ‹


Since the value of F(N) can be very large, return the result modulo 10â¹ + 7.

ðŸ”¹ Input Format

A single integer N

ðŸ”¹ Output Format

Print the value of F(N) modulo 10â¹ + 7

ðŸ”¹ Constraints
0 â‰¤ N â‰¤ 10^5

ðŸ”¹ Sample Input
5

ðŸ”¹ Sample Output
66

ðŸ” Sample Explanation

Using the recurrence:

F(0) = 1
F(1) = 1
F(2) = 1 + 7Ã—1 + (2/4) = 8
F(3) = 8 + 7Ã—1 + (3/4) = 15
F(4) = 15 + 7Ã—8 + (4/4) = 72
F(5) = 72 + 7Ã—15 + (5/4) = 66  (mod 10^9+7)

```python
def fellis(N):
    mod = 1000000007

    if N == 0 or N == 1:
        return 1

    f = [0] * (N + 1)
    f[0] = 1
    f[1] = 1

    for i in range(2, N + 1):
        f[i] = (f[i - 1] + 7 * f[i - 2] + (i // 4)) % mod

    return f[N]


# Input
N = int(input())

# Output
print(fellis(N))
```

34.You are given an integer N.

For every number M from 1 to N, define a function called starSum(M) as follows:

Convert M into a string

For each prefix of the string (from left to right), convert it back to an integer and add it to a sum

Formally, if M has k digits, then:

starSum(M) = int(first digit)
           + int(first 2 digits)
           + int(first 3 digits)
           + ...
           + int(all digits)


Your task is to count how many values of M (1 â‰¤ M â‰¤ N) satisfy:

starSum(M) > N

ðŸ”¹ Input Format

A single integer N

ðŸ”¹ Output Format

Print a single integer â€” the count of numbers M such that starSum(M) > N

ðŸ”¹ Constraints
1 â‰¤ N â‰¤ 10^5

ðŸ”¹ Sample Input
15

ðŸ”¹ Sample Output
6

```python
def star_sum(num):
    s = str(num)
    total = 0
    for i in range(1, len(s) + 1):
        total += int(s[:i])
    return total


N = int(input())
count = 0

for M in range(1, N + 1):
    if star_sum(M) > N:
        count += 1

print(count)
```

35.You are given an integer N and an integer array A of size N.

Your task is to find two different elements from the array such that:

Their sum is exactly 18

The first element is strictly greater than the second element

Among all such valid pairs, choose the pair whose product is maximum

If multiple pairs satisfy the conditions, print the pair with the maximum product.

ðŸ”¹ Input Format

An integer N â€” size of the array

N space-separated integers â€” elements of the array

ðŸ”¹ Output Format

Print two integers representing the required pair

Print them in the order:

larger_number smaller_number

ðŸ”¹ Constraints
2 â‰¤ N â‰¤ 100
-100 â‰¤ A[i] â‰¤ 100

ðŸ”¹ Sample Input
6
5 13 7 11 10 8

ðŸ”¹ Sample Output
13 5

```python
def find_pair(A):
    max_product = -10**18
    best_pair = [0, 0]

    for i in range(len(A)):
        for j in range(len(A)):
            if i != j and A[i] + A[j] == 18 and A[i] > A[j]:
                product = A[i] * A[j]
                if product > max_product:
                    max_product = product
                    best_pair[0] = A[i]
                    best_pair[1] = A[j]

    return best_pair


N = int(input())
A = list(map(int, input().split()))

result = find_pair(A)
print(result[0], result[1])
```

38.Flipkart Virus
Flipkart has been infected with a virus, the virus works this way: Each user that has been
infected with the virus is traced by a special algorithm and a hint that the virus gives. The
virus leaves a hint number N. Flipkart is able to identify the user ID of the virus by this N
number as the user ID works as a series : Each number in the series is the sum of the last
three numbers in the series. The first three numbers in the series are 0, 0, 1 always. Write a
program to identify the user ID infect based on N value checked from the logs of the system
Input
The input contains the N value left by the virus and found by engineers in the logs of the
system
Output
Print the userID of the infect user.
Example Input
11
Output
81

```python
N = int(input())

# Base cases
if N == 1 or N == 2:
    print(0)
elif N == 3:
    print(1)
else:
    a, b, c = 0, 0, 1  # first three numbers

    for _ in range(4, N + 1):
        d = a + b + c
        a, b, c = b, c, d

    print(c)
```

39.Amazon Cryptography
Amazon wants to apply cryptography to its barcode scanner printed on items. Each item
already has an item number, your job is to generate this cryptographed barcode number that
will be printed, which should follow the following rule:
Accept a crypto key C from the user, the item number replaced is then replaced with barcode
number such that each barcode digit is the Cth digit in the case if the result is less than 10.
Else, a character in general alphabets post 10.
See the input and output below to understand the solution.
Input format:
First input contains the order number and the 2nd input contains the crypto key C
Output format:
Print the order number
Example Input
46734
2
Output
68956

```python
order = input().strip()
C = int(input())

result = ""

for ch in order:
    d = int(ch)
    new_val = d + C

    if new_val < 10:
        result += str(new_val)
    else:
        # 10 -> 'A', 11 -> 'B', ...
        result += chr(ord('A') + new_val - 10)

print(result)
```

40.Count element of string
Write a program to count the number of digits, special characters, whitespaces, and
alphabets in a string. Print in count of each in same order, one per line
Input and Output
first line takes the input of a sentence
second line contains number of alphabets
third line contains number of digits
fourth line contains number of spaces
fifth line contains number of special characters
Sample
Input
Amcatuff@ #% 123
Output
Alphabets - 8
Digits - 3
Space - 2
Special Character - 3

```python
s = input()

alphabets = 0
digits = 0
spaces = 0
special = 0

for ch in s:
    if ('a' <= ch <= 'z') or ('A' <= ch <= 'Z'):
        alphabets += 1
    elif '0' <= ch <= '9':
        digits += 1
    elif ch == ' ':
        spaces += 1
    else:
        special += 1

print("Alphabets -", alphabets)
print("Digits -", digits)
print("Space -", spaces)
print("Special Character -", special)
```
