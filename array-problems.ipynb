```python
#1. Rotate the array N times 

def rotate_array(arr, n):
    if not arr:
        return arr
    n = n % len(arr)  # Handle cases where n is greater than array length
    return arr[-n:] + arr[:-n]

n = list(map(int, input("Enter array elements separated by space: ").split()))
k = int(input("Enter number of rotations: "))
rotated_array = rotate_array(n, k)
print("Rotated Array:", rotated_array)


```

```python
# Alternative method for rotation 

def rotate_array_alternative(arr, d):
     n = len(arr)
  
    # Repeat the rotation d times
     for i in range(d):
      
        # Left rotate the array by one position
        first = arr[0]
        for j in range(n - 1):
            arr[j] = arr[j + 1]
        arr[n - 1] = first
     return arr
n = list(map(int, input("Enter array elements separated by space: ").split()))
k = int(input("Enter number of rotations: "))
rotated_array = rotate_array(n, k)
print("Rotated Array:", rotated_array)
```

```python
# Best time to buy and sell. 

li = [7,1,5,3,6,4]

buy = li[0]
sell = 0
profit = 0

for i in range(1,len(li)):
    if li[i] < buy:
        buy = li[i]
    sell = li[i] - buy 
    profit= max(profit , sell)
print(profit)
```

```python
# 3.longest alternating length
def longest_alternating_length(s):
    if not s:
        return 0

    current_len = 1
    max_len = 1

    for i in range(1, len(s)):
        if s[i] != s[i - 1]:
            current_len += 1
            max_len = max(max_len, current_len)
        else:
            current_len = 1

    return max_len


s = "1010110"
print(longest_alternating_length(s))  # Output: 5
```
